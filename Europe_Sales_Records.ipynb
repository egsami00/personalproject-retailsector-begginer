```python
mapping = {1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5: 'May', 6:'Jun', 7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'}
df['order_day_of_the_month'] = df['Order Date'].dt.day
df['order_by_day_of_the_week'] = df['Order Date'].dt.dayofweek
df['order_month_of_the_year'] = df['Order Date'].dt.month.map(mapping)
df['order_year'] = df['Order Date'].dt.year
df['order_quarter'] = df['Order Date'].dt.quarter

df.head()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x='order_year', y='Total Revenue', hue='Sales Channel', data=df)
plt.title('Total Revenue by Year and Sales Channel (Mean)')
plt.xlabel('Year')
plt.ylabel('Total Revenue')
plt.show()
```

```python
plt.figure(figsize=(15,10))
color = ['#FF653F', '#FFC85C']
sns.barplot(x='order_year', y='Total Revenue', hue='Sales Channel', estimator=np.sum, palette=color, data=df)
plt.title('Total Revenue by Year and Sales Channel')
plt.xlabel('Year')
plt.ylabel('Total Revenue')
plt.show()
```
